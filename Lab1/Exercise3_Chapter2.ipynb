# Building a Robot Simulator in 2D

```python
# Perfect sensor
def sense(x):
    return x
```

### Euler integration

```python
def simulate(Δt, x, u):
    x += Δt * u
    return x
```

### Linear Control policy

```python
def control(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    ux = -sin(t)
    uy = cos(t * 2)
    return array([ux, uy])
```

### Line Trajectory

```python
tf = 10.    # Time final
Δt = 0.1    # Time step
time = linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
x = array([1., 0.])
x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y)
    x = simulate(Δt, x, u)
    x_log.append(copy(x))

x_log = array(x_log)
```

```python
grid()
plot(x_log[:,0], x_log[:,1])
```
